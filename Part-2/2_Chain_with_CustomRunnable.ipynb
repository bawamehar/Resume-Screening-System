```python
from  langchain_openai import ChatOpenAI
import os 
from dotenv import load_dotenv

# This function will load all the variables from the .env file and will 
# make them available in the os.environ dictionary (env variables)
load_dotenv() 

if os.environ.get("OPENAI_API_KEY"):
    print("Bro API KEY Variable exists")
else:
    raise ValueError("OPENAI_API_KEY not found")

from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from  langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0)
```

# Chain with Custom Runnable

```python
# Task 1 Prompt

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "you are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# Task 2 LLM
llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0)
```

```python
# Task 3 Str Parser
str_parser = StrOutputParser()
```

```python
# Task 4 Custom Runnable
from langchain_core.runnables import RunnableLambda

# we cannot pass or run a python function directly in the chain, 
# we need to convert/wrap it in a RunnableLambda

def dict_maker(text: str) -> dict:
    return {"text" : text}

dict_maker_runnable = RunnableLambda(dict_maker)
```

```python
# Task 5 Template for post

prompt_post = ChatPromptTemplate.from_messages(
    messages=[
        ("system", "you are a social media post generator"),
        ("human", "create a post for the following text for linkedin: {text}")
    ]

)
```

```python
# Task 6 Generate the post

llm_openai = ChatOpenAI(model="gpt-5-mini",temperature=0)
```

```python
#Task 7 Str parser

str_parser_post = StrOutputParser()
```

```python
#chain

chain = prompt_template | llm_openai | str_parser | dict_maker_runnable | prompt_post | llm_openai | str_parser_post
chain.invoke({"input" : "What is the capital of France"})
```
